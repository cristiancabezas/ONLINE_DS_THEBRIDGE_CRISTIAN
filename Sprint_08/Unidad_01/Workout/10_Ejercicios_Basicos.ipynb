### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
plt.style.use('seaborn-v0_8-whitegrid')
x = np.linspace(0,10,100)
```

```python
fig = plt.figure()
plt.subplot(1,2,1)
plt.plot(x, np.sin(x))
plt.title('Seno')
plt.yticks(np.arange(-1, 1.25, 0.25))
plt.xticks(np.arange(0,11, 2.5))
plt.ylim(-1.10, 1.10)
plt.grid(False)

plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
plt.title('Coseno')
plt.yticks(np.arange(-1, 1.25, 0.25))
plt.xticks(np.arange(0,11, 2.5))
plt.ylim(-1.10, 1.10)
plt.grid(False);
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
fig, ax = plt.subplots(3,1)
ax[0].plot(x, np.cos(x), color = 'red')
ax[1].plot(x, np.sin(x), color = 'green')
ax[0].set_yticks(np.arange(-1, 2, 1))
ax[1].set_yticks(np.arange(-1, 2, 1))
ax[0].set_ylim(-1.10, 1.10)
ax[1].set_ylim(-1.10, 1.10)

ax[2].plot(x, np.tan(x))
ax[2].set_yticks(np.arange(-40,21, 20))
ax[2].set_ylim(-40,20)
ax[0].grid(False)
ax[1].grid(False)
ax[2].grid(False)
plt.show()
```


### Ejercicio 3: Ajuste de Colores y Estilos de Lnea  

Modifica la primera figura para que se muestre de la siguiente manera

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperblico y coseno hiperblico se obtienen as:
* np.sinh, seno hiperblico;
* np.cosh, coseno hiperblico;


```python
fig, ax = plt.subplots(2,1, figsize = (10, 9))
ax[0].plot(x, np.sinh(x), color = 'red')
ax[0].set_yticks(np.arange(0,10001, 2500))
ax[1].plot(x, np.cosh(x), color = 'blue')
ax[1].set_yticks(np.arange(0,10001, 2500))
ax[0].set_title('Seno Hiperblico')
ax[0].set_ylabel('sinh(x)')
ax[0].set_xlabel('x')
ax[1].set_title('Coseno Hiperblico')
ax[1].set_ylabel('cosh(x)')
ax[1].set_xlabel('x')
ax[0].grid(False)
ax[1].grid(False)
ax[0].set_ylim(-200)
ax[1].set_ylim(-200)
plt.show()
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosicin (tal y como ves en la figura), emplea `subplots_adjust` y los parmetros hspace y wspace.

```python
fig, ax = plt.subplots(2,2)
fig.subplots_adjust(hspace = 0.3, wspace = 0.3)

ax[0,0].grid(False)
ax[0,0].plot(x, np.sin(x))
ax[0,0].set_ylim(-1.05, 1.10)

ax[0,1].grid(False)
ax[0,1].plot(x, np.cos(x))
ax[0,1].set_ylim(-1.05, 1.10)

ax[1,0].grid(False)
ax[1,0].plot(x, np.tan(x))
ax[1,0].set_ylim(-40, 20)


ax[1,1].grid(False)
ax[1,1].plot(x, np.sqrt(x))

plt.show()
```


### Ejercicio 6: 

Reproduce la figura siguiente empleando slo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros grficos (fila 1). Luego usa el estilo con mtodos (objetos `axes`) para aadir el tercer grfico.

```python
fig, ax = plt.subplots(2,1)
fig.subplots_adjust(hspace = 0.3)
plt.sca(ax[0]) #Para elegir el sitio donde ponerlo
plt.plot(x, np.cos(x), label = 'coseno')
plt.plot(x, np.sqrt(x), color = 'orange', label = 'raiz cuadrada')
plt.title('Raz Cuadrada y coseno')
plt.legend()
plt.yticks(np.arange(-1,4,1))
plt.grid(False)

ax[1].plot(x, np.sin(x))
ax[1].set_title('Seno')
ax[1].grid(False)
ax[1].set_ylim(-1.05, 1.10)

plt.show()
```
